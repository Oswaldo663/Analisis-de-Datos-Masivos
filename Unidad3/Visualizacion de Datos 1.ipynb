```python
import pandas as pd
import numpy as np
```

### Paso 1. Carga, limpiesa y transformacion de datos

```python
df_canada = pd.read_excel('Canada.xlsx', sheet_name = 'Canada by Citizenship',
                       skiprows =range(20),
                       skipfooter=2 )
df_canada
```

```python
print (df_canada.shape)
```

```python
#borrar columnas
df_canada.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'],
              axis=1, inplace=True)
df_canada.head(5)
```

```python
df_canada.rename(columns = {'OdName':'Country',
                           'AreaName':'Continent',
                           'RegName':'Region'}, inplace=True)
df_canada
```

```python
#Verificar que todos los tipos  de columnas sean string
all(isinstance(column, str) for column in df_canada.columns)
```

```python
#cambiar todas las columnas a tpo string
df_canada.columns = list(map(str, df_canada.columns))
```

```python
#Verificar que todos los tipos  de columnas sean string
all(isinstance(column, str) for column in df_canada.columns)
```

```python
df_canada.set_index('Country', inplace=True)
df_canada.head()
```

```python
#agregamos la columna total sumando la imigracion 
#de todos los años por pais
df_canada['Total'] =df_canada.sum(axis=1)
df_canada.head()
```

```python
#mapeo para convertir a tipo string
year = list (map(str, range(1980, 2014)))
year
```

### Parte 2.- Visualizacion de DAtos con matplotlib

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

#### Graficos de Area

```python
#Ordenar en este caso por la columna total
df_canada.sort_values(['Total'], ascending=False, 
                      axis=0, inplace=True)
df_canada.head(10)
```

```python
df_top5 = df_canada.head()
df_top5
```

```python
#Cambio de nombres a columnas y años a renglones para poder graficar
df_top5 = df_top5[year].transpose()
df_top5
```

```python
df_top5.index = df_top5.index.map(int)
df_top5
```

```python
df_top5.plot(kind='area',
            stacked=False,
            figsize=(20,10) )
plt.title('Tendencias de Inmigracion de los primeros 5 paises a Canada')
plt.ylabel('Numero de inmigrantes')
plt.xlabel('Años')
plt.show()
```

```python
#Transparencia
df_top5.plot(kind='area',
             alpha=0.25,
            stacked=False,
            figsize=(20,10) )
plt.title('Tendencias de Inmigracion de los primeros 5 paises a Canada')
plt.ylabel('Numero de inmigrantes')
plt.xlabel('Años')
plt.show()
```

```python
ax = df_top5.plot(kind='area',
             alpha=0.25,
            figsize=(20,10) )
ax.set_title('Tendencias de inmigracion de los 5 primeros paises a canada')
ax.set_ylabel('Numero de inmigrantes')
ax.set_xlabel('Años')
```

#### Graficos de Barras

```python
df_chile =df_canada.loc['Chile', year]
df_chile.head()
```

```python
#en los diagramas de barras el eje x se usa para etiquetar
#y la longitud de la barra corresponde al valor de la variable 
df_chile.plot(kind='barh', figsize=(10,10) )
plt.xlabel('Numero de inmigrantes')
plt.ylabel('Años')
plt.title('Inmigrantes chilenos a canada desde 1980 al 2013')
for index, value in enumerate(df_chile):
    label = format (int(value), ',')
    plt.annotate(label, xy=(value, index), color='black')
plt.show()
```

```python
df_mexico =df_canada.loc['Mexico', year]
df_mexico.head()
```

```python
df_mexico.plot(kind='bar', figsize=(10,6) )
plt.xlabel('Años')
plt.ylabel('Numero de inmigrantes')
plt.title('Inmigrantes Mexicanos a canada desde 1980 al 2013')
plt.show()
```
