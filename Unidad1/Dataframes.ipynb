## Introduccion a dataframes <a id=id1> <a/>

```python
import pandas as pd
```

```python
Alumno1 = pd.Series({'Nombre':'Juan', 'Fecha Nacimiento':'10-02-2000', 'Carrera':'IINF', 'Edad':20})
Alumno2 = pd.Series({'Nombre':'Karla', 'Fecha Nacimiento':'18-05-2001', 'Carrera':'ISIC', 'Edad':19})
Alumno3 = pd.Series({'Nombre':'Pedro', 'Fecha Nacimiento':'15-07-1998', 'Carrera':'IINF', 'Edad':22})
Alumno4 = pd.Series({'Nombre':'Maria', 'Fecha Nacimiento':'10-12-1999', 'Carrera':'ITIC', 'Edad':19})
```

```python
Alumnos = pd.DataFrame([Alumno1, Alumno2, Alumno3, Alumno4])
Alumnos
```

## Load CSV as DataFrames <a id=id5> <a/>

```python
df = pd.read_csv('olympics.csv')
df
```

```python
df.head()
```

```python
df = pd.read_csv('olympics.csv', index_col = 0, skiprows = 1)
df.head()
```

```python
df.columns
```

```python
for col in df.columns:
    if col[:2] == '01':
        df.rename(columns = {col:'Gold' + col[4:]}, inplace = True)
    if col[:2] == '02':
        df.rename(columns = {col:'Silver' + col[4:]}, inplace = True)
    if col[:2] == '03':
        df.rename(columns = {col:'Bronze' + col[4:]}, inplace = True)
df.head()
```

## Busquedas en un DataFrame <a id=id6> <a/>

```python
df['Gold'] > 0
df['Gold'].head(10) > 0
```

```python
ganadores_oro = df.where(df['Gold'] > 0)
ganadores_oro
```

```python
ganadores_oro['Gold'].count()
```

```python
ganadores_oro = ganadores_oro.dropna()
ganadores_oro
```

```python
ganadores_oro2 = df[df['Gold'] > 0]
ganadores_oro2
```

```python
ganadores_oro3 = df[(df['Gold'] > 0) & (df['Gold.1'] > 0)]
ganadores_oro3
```
